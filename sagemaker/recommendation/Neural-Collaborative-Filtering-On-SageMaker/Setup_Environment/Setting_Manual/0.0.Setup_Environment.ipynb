# [모듈 0.0] 역할 및 권한 셋업 가이드
- 이미 이 워크샵에서 아래의 권한 및 역할 설정에 대한 참고 자료 입니다. 수동으로 적용시 참고 하세요.


# 1. SageMaker Notebook Role
- 이 노트북을 실행하는 role 은 아래와 같은 역할에 대한 권한이 필요 합니다.
    - IAMFullAccess
    - AmazonS3FullAccess
    - AWSCodeBuildAdminAccess
    - AWSCodeCommitFullAccess
    - AWSCodePipelineFullAccess
    - AmazonSageMakerFullAccess
    - AmazonEventBridgeFullAccess
    - AWSLambda_FullAccess # 세이지 메이커 파이프라인에서 람다 스템 실행
- 필요한 신뢰 정책    
    - Trusted Entity
        - "codebuild.amazonaws.com",
        - "sagemaker.amazonaws.com"
        ```json
        {
    "Version": "2012-10-17",
    "Statement": [
        {
            "Effect": "Allow",
            "Principal": {
                "Service": [
                    "sagemaker.amazonaws.com",
                    "lambda.amazonaws.com"                    
                ]
            },
            "Action": "sts:AssumeRole"
        }
    ]
}
        ```                
    


# 2. codebuild 실행 Role 생성 필요
- codebuild 를 실행하는 역할이 필요 합니다. 이 역할은 아래와 같은 권한이 필요 합니다.
- codebuild_service_role:
    - CodebuildFullAccess
    - AWSCodeCommitFullAccess
    - CloudWatchFullAccess    
    - AmazonS3FullAccess    
    - AmazonSageMakerFullAccess
    - AWSLambda_FullAccess # 세이지 메이커 파이프라인에서 람다 스템 실행
- 필요한 신뢰 정책    
    - Trusted Entity
        - "codebuild.amazonaws.com",
        - "sagemaker.amazonaws.com"
        ```json
        {
    "Version": "2012-10-17",
    "Statement": [
        {
            "Effect": "Allow",
            "Principal": {
                "Service": [
                    "codebuild.amazonaws.com",
                    "sagemaker.amazonaws.com",
                    "lambda.amazonaws.com"                    
                ]
            },
            "Action": "sts:AssumeRole"
        }
    ]
}
        ```
