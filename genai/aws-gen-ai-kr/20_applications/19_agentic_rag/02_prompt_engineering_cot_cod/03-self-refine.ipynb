## ì¼ë°˜ ìƒ˜í”Œì½”ë“œ

```python
import boto3

def generate_code_with_bedrock(query, model_id="amazon.nova-lite-v1:0", region_name="us-east-1"):
    # Bedrock í´ë¼ì´ì–¸íŠ¸ ìƒì„±
    client = boto3.client("bedrock-runtime", region_name=region_name)
    
    # ì½”ë“œ ìƒì„±ì„ ìœ„í•œ ì‹œìŠ¤í…œ í”„ë¡¬í”„íŠ¸
    system_prompt = """
    ë‹¹ì‹ ì€ ìœ ëŠ¥í•œ ì½”ë“œ ìƒì„± ë¹„ì„œì…ë‹ˆë‹¤. ì‚¬ìš©ìì˜ ìš”ì²­ì— ë”°ë¼ ê¹”ë”í•˜ê³  íš¨ìœ¨ì ì¸ ì½”ë“œë¥¼ ì‘ì„±í•´ì£¼ì„¸ìš”.
    ì½”ë“œì— ëŒ€í•œ ì„¤ëª…ì€ ê°„ê²°í•˜ê²Œ ì£¼ì„ìœ¼ë¡œ í¬í•¨ì‹œí‚¤ê³ , ê°€ëŠ¥í•œ ëª¨ë²” ì‚¬ë¡€ë¥¼ ë”°ë¥´ì„¸ìš”.
    """
    
    # Bedrock ëª¨ë¸ í˜¸ì¶œ
    response = client.converse(
        modelId=model_id,
        messages=[
            {
                "role": "user",
                "content": [{"text": query}]
            }
        ],
        system=[{"text": system_prompt}],
        inferenceConfig={
            "maxTokens": 500,
            "temperature": 0.7,
            "topP": 0.9
        }
    )
    
    # ì‘ë‹µ ì¶”ì¶œ
    return response["output"]["message"]["content"][0]["text"]

    
```

```python
import boto3
import json

def self_refine_with_bedrock(query, model_id="amazon.nova-lite-v1:0", region_name="us-east-1", iterations=1):
    # Bedrock í´ë¼ì´ì–¸íŠ¸ ìƒì„±
    client = boto3.client("bedrock-runtime", region_name=region_name)
    
    # ì´ˆê¸° ì‘ë‹µ ìƒì„±
    initial_response = generate_response(client, model_id, query)
    current_response = initial_response
    
    # ì§€ì •ëœ íšŸìˆ˜ë§Œí¼ ìê¸° ê°œì„  ë°˜ë³µ
    for i in range(iterations):
        # í˜„ì¬ ì‘ë‹µì— ëŒ€í•œ í”¼ë“œë°± ìƒì„±
        feedback = generate_feedback(client, model_id, query, current_response)
        
        # í”¼ë“œë°±ì„ ë°”íƒ•ìœ¼ë¡œ ê°œì„ ëœ ì‘ë‹µ ìƒì„±
        improved_response = generate_improved_response(client, model_id, query, current_response, feedback)
        current_response = improved_response
        
    return {
        "initial_response": initial_response,
        "final_response": current_response
    }

def generate_response(client, model_id, query):
    # ì´ˆê¸° ì‘ë‹µ ìƒì„±
    system_prompt = "ë‹¹ì‹ ì€ ìœ ìš©í•œ AI ë¹„ì„œì…ë‹ˆë‹¤. ì‚¬ìš©ìì˜ ì§ˆë¬¸ì— ì •í™•í•˜ê³  ë„ì›€ì´ ë˜ëŠ” ë‹µë³€ì„ ì œê³µí•˜ì„¸ìš”."
    
    response = client.converse(
        modelId=model_id,
        messages=[
            {
                "role": "user",
                "content": [{"text": query}]
            }
        ],
        system=[{"text": system_prompt}],
        inferenceConfig={
            "maxTokens": 500,
            "temperature": 0.7,
            "topP": 0.9
        }
    )
    
    return response["output"]["message"]["content"][0]["text"]

def generate_feedback(client, model_id, query, response):
    # ì‘ë‹µì— ëŒ€í•œ í”¼ë“œë°± ìƒì„±
    feedback_prompt = f"""
    ë‹¤ìŒ ì§ˆë¬¸ê³¼ ì‘ë‹µì„ í‰ê°€í•˜ê³ , ê°œì„ í•  ì ì„ ì œì•ˆí•´ì£¼ì„¸ìš”:
    
    ì§ˆë¬¸: {query}
    
    ì‘ë‹µ: {response}
    
    ì´ ì‘ë‹µì˜ ê°•ì ê³¼ ì•½ì ì„ ë¶„ì„í•˜ê³ , êµ¬ì²´ì ì¸ ê°œì„  ë°©ì•ˆì„ ì œì‹œí•´ì£¼ì„¸ìš”.
    """
    
    feedback_response = client.converse(
        modelId=model_id,
        messages=[
            {
                "role": "user",
                "content": [{"text": feedback_prompt}]
            }
        ],
        inferenceConfig={
            "maxTokens": 500,
            "temperature": 0.7,
            "topP": 0.9
        }
    )
    
    return feedback_response["output"]["message"]["content"][0]["text"]

def generate_improved_response(client, model_id, query, original_response, feedback):
    # í”¼ë“œë°±ì„ ë°”íƒ•ìœ¼ë¡œ ê°œì„ ëœ ì‘ë‹µ ìƒì„±
    improvement_prompt = f"""
    ë‹¤ìŒ ì§ˆë¬¸ì— ëŒ€í•œ ì›ë˜ ì‘ë‹µê³¼ í”¼ë“œë°±ì„ ê³ ë ¤í•˜ì—¬ ê°œì„ ëœ ì‘ë‹µì„ ì‘ì„±í•´ì£¼ì„¸ìš”:
    
    ì§ˆë¬¸: {query}
    
    ì›ë˜ ì‘ë‹µ: {original_response}
    
    í”¼ë“œë°±: {feedback}
    
    ìœ„ í”¼ë“œë°±ì„ ë°˜ì˜í•˜ì—¬ ê°œì„ ëœ ì‘ë‹µì„ ì‘ì„±í•´ì£¼ì„¸ìš”.
    """
    
    improved_response = client.converse(
        modelId=model_id,
        messages=[
            {
                "role": "user",
                "content": [{"text": improvement_prompt}]
            }
        ],
        inferenceConfig={
            "maxTokens": 500,
            "temperature": 0.7,
            "topP": 0.9
        }
    )
    
    return improved_response["output"]["message"]["content"][0]["text"]
```
